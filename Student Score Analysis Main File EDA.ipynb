#Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = "Gender", data = df, palette = ['purple','blue'], hue = "Gender")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Gender Distribution")
plt.show()
```

#From the above chart we have analysed that:
#the number of females in the data is more than the number of males.

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

#From the above chart we have concluded that the education of the parents have a good impact on their scores.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

#From the above chart we have concluded that:
#There is no/negligible impact on the students score due to their parents marital status.

### Percentage Distribuition of Ethnic Groups

```python
ax = sns.countplot(data = df, x ="EthnicGroup", palette = ['red', 'green', 'blue', 'darkorange', 'purple'], hue = "EthnicGroup")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
ax.bar_label(ax.containers[4])
plt.show()
```

```python
groupA= df.loc[(df["EthnicGroup"] == "group A")].count()
groupB= df.loc[(df["EthnicGroup"] == "group B")].count()
groupC= df.loc[(df["EthnicGroup"] == "group C")].count()
groupD= df.loc[(df["EthnicGroup"] == "group D")].count()
groupE= df.loc[(df["EthnicGroup"] == "group E")].count()

colors = ["blue", "green", "red", "darkorange", "purple"]
plt.figure(figsize = (5,5))
l =["groupA", "groupB", "groupC", "groupD", "groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"],  groupD["EthnicGroup"],  groupE["EthnicGroup"]]
plt.pie(mlist, labels =l , autopct = "%1.2f%%", colors=colors, wedgeprops={"edgecolor": "black", "linewidth": 1.5}, shadow = True)
plt.title("Distribution of Ethnic Groups")
plt.show()
print(mlist)
```
